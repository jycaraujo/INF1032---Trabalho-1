```python
import pandas as pd

csv_file = "discentes.csv"
discentes_raw = pd.read_csv(csv_file)
discentes_data = discentes_raw
discentes_data.shape
discentes_data.rename(columns = {'nível': 'nivel'}, inplace=True)
discentes_data.head()
```

```python
csv_file = "trabalhos.csv"
trabalhos_raw = pd.read_csv(csv_file)
trabalhos_data = trabalhos_raw
trabalhos_data.shape
# trabalhos_data.rename(columns = {'nível': 'nivel'}, inplace=True)
trabalhos_data.head()
```

```python
# Descobrir quais colunas sao comuns as duas tabelas
common_columns = []
for column_name in discentes_data.columns:
    if(column_name in trabalhos_data):
        common_columns.append(column_name)
common_columns
```

```python
# Utilizando variaveis auxiliares para mexer com as tabelas
aux_discentes_data = discentes_data.copy()
aux_trabalhos_data = trabalhos_data.copy()

# Manter somente as colunas necessárias
to_drop_in_discentes = []
for column_name in aux_discentes_data.columns:
    if(column_name not in common_columns and column_name != 'nivel'):
        to_drop_in_discentes.append(column_name)

aux_discentes_data = aux_discentes_data.drop(to_drop_in_discentes, 1)

to_drop_in_trabalhos = []
for column_name in aux_trabalhos_data.columns:
    if(column_name not in common_columns and column_name != 'data_defesa'):
        to_drop_in_trabalhos.append(column_name)

aux_trabalhos_data = aux_trabalhos_data.drop(to_drop_in_trabalhos, 1)
```

```python
# Preparando os dados
possible_ies = {}
possible_programas = {}
possible_orientadores = {}

def match_strings(str1, str2, error):
    diffs = 0
    aux_str1 = str1.replace(" ", "")
    aux_str2 = str2.replace(" ", "")
    for a, b in zip(aux_str1, aux_str2):
        if a != b:
            diffs += 1
            if(diffs > error):
                return 0
    return 1

def treat_string(c, column):
#     print(column)
    if(not isinstance(c, str)):
        return c
    else:
        if(column == 'ies'):
            possible_group = possible_ies
            error = 10
        elif(column == 'programa'):
            possible_group = possible_programas
            error = 2
        elif(column == 'orientador'):
            possible_group = possible_orientadores
            error = 2

        for key in possible_group:
            if(c in possible_group[key]):
                return key
                break
            else:
                for possible_name in possible_group[key]:
                    if(match_strings(c, possible_name, error)):
                        possible_group[key].append(c);
                        return key
                        break
        possible_group[c] = [c]
        return c
    
for column in aux_discentes_data:
    if(column in common_columns):
        aux_discentes_data[column] = aux_discentes_data[column].apply(treat_string, args=([column]))

for column in aux_trabalhos_data:
    if(column in common_columns):
        aux_trabalhos_data[column] = aux_trabalhos_data[column].apply(treat_string, args=([column]))
```

```python
from bokeh.io import show, output_notebook
from bokeh.models import ColumnDataSource, FactorRange
from bokeh.plotting import figure

output_notebook()

cursos = ['curso1', 'curso2', 'curso3', 'curso4', 'curso5', 'curso6']
alunos = ['Mestrado', 'Doutorado']

data = {'cursos' : cursos,
        'Mestrado'   : [2, 1, 4, 3, 2, 4],
        'Doutorado'   : [5, 3, 3, 2, 4, 6]}

# this creates [ ("Apples", "2015"), ("Apples", "2016"), ("Apples", "2017"), ("Pears", "2015), ... ]
x = [ (curso, aluno) for curso in cursos for aluno in alunos ]
counts = sum(zip(data['Mestrado'], data['Doutorado']), ()) # like an hstack

source = ColumnDataSource(data=dict(x=x, counts=counts))

p = figure(x_range=FactorRange(*x), plot_height=250, title="Alunos de Mestrado por Curso",
           toolbar_location=None, tools="")

p.vbar(x='x', top='counts', width=0.9, source=source)

p.y_range.start = 0
p.x_range.range_padding = 0.1
p.xaxis.major_label_orientation = 1
p.xgrid.grid_line_color = None

show(p)
```
